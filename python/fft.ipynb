```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from scipy.fftpack import fft
```

```python
sample_rate = 1024
N = (2 - 0) * sample_rate
```

```python
freq1 = 60
magnitude1 = 25
freq2 = 270
magnitude2 = 2
time = np.linspace(0, N*1/freq1, N)
waveform1 = magnitude1 * np.sin(2 * pi * freq1 * time)
waveform2 = magnitude2 * np.sin(2 * pi * freq2 * time)
```

```python
normal = np.random.normal(0, 3, N)
time_data = waveform1 + waveform2 + normal
```

```python
plt.plot(time, time_data)
plt.title('Time Domain Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
# frequency = np.linspace(0, 500, 1024)
freq_data = fft(time_data)
y = 2/N * np.abs(freq_data[0:int(N/2)])
frequency = np.fft.fftfreq(len(y), 1/N)
```

```python
plt.plot(frequency[0:len(frequency)//2], y[0:len(frequency)//2])
plt.title('Frequency domain Signal')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.show()
```
